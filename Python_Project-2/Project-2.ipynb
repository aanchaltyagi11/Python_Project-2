```python
from tkinter import *

Calculator=Tk()
Calculator.title("Mini_Calculator")
Calculator.resizable(width=False, height=False)
Calculator.geometry("400x490+550+250")
Calculator.configure(bg="skyblue")
Calculator.iconbitmap(r'C:\Users\Aanchal Tyagi\Desktop\Python_Project-2\images')

# Button function
def calculate():
    bs = int(Basic_Salary_entry.get())
    ot = int(Over_Time_entry.get())
    hra = int(HRA_entry.get())
    co = int(Conveyancey_entry.get())
    medi = int(Medical_entry.get())
    total=(bs + ot + hra + co + medi)
    Label(Calculator, text=int(total), font=("Times New Roman",12,'bold'), bg="skyblue",fg="black").place(relx=0.4,rely=0.58)
    b=Label(Calculator, text="Total", font=("Times New Roman",12,'bold'), bg="skyblue",fg="black")
    b.place(relx=0.1,rely=0.58)

def clear_entry():
    Basic_Salary_entry.delete(0,END)
    Over_Time_entry.delete(0,END)
    HRA_entry.delete(0,END)
    Conveyancey_entry.delete(0,END)
    Medical_entry.delete(0,END)
    
    
# Heading
Label(Calculator, text="Salary Calculation", font=("Calibri",19,'bold'), bg="lightpink",fg="black").pack(fill=X)

# Lable
b1=Label(Calculator, text="Basic Salary", font=("Times New Roman",12), bg="skyblue",fg="black")
b2=Label(Calculator, text="Over Time", font=("Times New Roman",12), bg="skyblue",fg="black")
b3=Label(Calculator, text="HRA", font=("Times New Roman",12), bg="skyblue",fg="black")
b4=Label(Calculator, text="Conveyance", font=("Times New Roman",12), bg="skyblue",fg="black")
b5=Label(Calculator, text="Medical", font=("Times New Roman",12), bg="skyblue",fg="black")


b1.place(relx=0.1,rely=0.20)
b2.place(relx=0.1,rely=0.27)
b3.place(relx=0.1,rely=0.34)
b4.place(relx=0.1,rely=0.41)
b5.place(relx=0.1,rely=0.49)


# Entry Box
Basic_Salary_entry = Entry(Calculator, bd=2,width=15, font=("Times New Roman",11))
Over_Time_entry = Entry(Calculator, bd=2,width=15, font=("Times New Roman",11))
HRA_entry = Entry(Calculator, bd=2,width=15, font=("Times New Roman",11))
Conveyancey_entry = Entry(Calculator, bd=2,width=15, font=("Times New Roman",11))
Medical_entry = Entry(Calculator, bd=2,width=15, font=("Times New Roman",11))


Basic_Salary_entry.place(relx=0.4,rely=0.20)
Over_Time_entry.place(relx=0.4,rely=0.27)
HRA_entry.place(relx=0.4,rely=0.34)
Conveyancey_entry.place(relx=0.4,rely=0.41)
Medical_entry.place(relx=0.4,rely=0.49)


# Button
Button(Calculator, text="Calculate", font=("Arial",12,'bold'), bd=8,bg="black",fg="white", command=calculate).place(relx=0.1,rely=0.82)
Button(Calculator, text="Clear", font=("Arial",12,'bold'), bd=8,bg="red",fg="white", width=8, command=clear_entry).place(relx=0.65,rely=0.82)




Calculator.mainloop()
```
